```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/think_school_non_podcast_dataset.csv')
```

```python
df.head()
```

```python
df['published_at'] = pd.to_datetime(df['published_at'], errors='coerce')
df['published_date'] = df['published_at'].dt.date
```

```python
df.head()
```

```python
df = df.drop(columns=['video_id', 'published_at', 'duration', 'podcast', 'is_short_strict'])
```

```python
df.head()
```

```python
df.shape
```

```python
df = df[df['duration_minutes'] >= 2.5]
```

```python
df.shape
```

```python
df.head()
```

```python
df['duration_minutes'] = df['duration_minutes'].round(-2)
```

```python
df.head()
```

```python
df = df.drop(columns=['duration_minutes'])
```

```python
df.head()
```

```python
import matplotlib.ticker as mtick
```

```python
df['published_date'] = pd.to_datetime(df['published_date'], errors='coerce')
df['year_month'] = df['published_date'].dt.to_period('M').astype(str)
df['year'] = df['published_date'].dt.year
df['month_name'] = df['published_date'].dt.month_name()
```

### Average month+year views per video 

```python
ym_views = df.groupby('year_month').agg(
    total_views=('view_count','sum'),
    n_videos=('title','count')
).reset_index()
ym_views['avg_views'] = ym_views['total_views'] / ym_views['n_videos']
ym_views = ym_views.sort_values('year_month')

plt.figure(figsize=(12,5))
plt.plot(ym_views['year_month'], ym_views['avg_views'], marker='o')
plt.title("Average Views per Case Study video")
plt.xlabel("Year-Month")
plt.ylabel("Average Views")
plt.xticks(rotation=90)
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.tight_layout()
plt.show()
```

### Average month + year likes per video

```python
ym_likes = df.groupby('year_month').agg(
    total_likes=('like_count','sum'),
    n_videos=('title','count')
).reset_index()
ym_likes['avg_likes'] = ym_likes['total_likes'] / ym_likes['n_videos']
ym_likes = ym_likes.sort_values('year_month')

plt.figure(figsize=(12,5))
plt.plot(ym_likes['year_month'], ym_likes['avg_likes'], marker='o')
plt.title("Average Likes per Video")
plt.xlabel("Year-Month")
plt.ylabel("Average Likes")
plt.xticks(rotation=90)
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.tight_layout()
plt.show()
```

### Average month + year comments per video

```python
ym_comments = df.groupby('year_month').agg(
    total_comments=('comment_count','sum'),
    n_videos=('title','count')
).reset_index()
ym_comments['avg_comments'] = ym_comments['total_comments'] / ym_comments['n_videos']
ym_comments = ym_comments.sort_values('year_month')

plt.figure(figsize=(12,5))
plt.plot(ym_comments['year_month'], ym_comments['avg_comments'], marker='o')
plt.title("Average Comments per Video")
plt.xlabel("Year-Month")
plt.ylabel("Average Comments")
plt.xticks(rotation=90)
plt.gca().yaxis.set_major_formatter(mtick.StrMethodFormatter('{x:,.0f}'))
plt.tight_layout()
plt.show()
```

### Top 5 videos by views

```python
top5 = df.sort_values('view_count', ascending=False).head(5)
display(top5[['title','view_count','like_count','comment_count','published_date']])
```

### Heatmap — Average Views per Month-Year

```python
avg_view_hm = df.groupby(['year','month_name']).agg(
    total_views=('view_count','sum'),
    n_videos=('title','count')
).reset_index()
avg_view_hm['avg_views'] = avg_view_hm['total_views'] / avg_view_hm['n_videos']

# pivot
all_months = ['January','February','March','April','May','June','July','August','September','October','November','December']
pivot_views = avg_view_hm.pivot_table(values='avg_views', index='year', columns='month_name').reindex(columns=all_months)

plt.figure(figsize=(14,6))
sns.heatmap(pivot_views, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Heatmap: Average Views per Video")
plt.xlabel("Month")
plt.ylabel("Year")
plt.tight_layout()
plt.show()
```

### Heatmap — Average Likes per Month-Year

```python
avg_like_hm = df.groupby(['year','month_name']).agg(
    total_likes=('like_count','sum'),
    n_videos=('title','count')
).reset_index()
avg_like_hm['avg_likes'] = avg_like_hm['total_likes'] / avg_like_hm['n_videos']

pivot_likes = avg_like_hm.pivot_table(values='avg_likes', index='year', columns='month_name').reindex(columns=all_months)

plt.figure(figsize=(14,6))
sns.heatmap(pivot_likes, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Heatmap: Average Likes per Video")
plt.xlabel("Month")
plt.ylabel("Year")
plt.tight_layout()
plt.show()
```
